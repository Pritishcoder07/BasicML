## Summary of NumPy Operations

This notebook provides an introduction to fundamental NumPy array creation and manipulation techniques:

*   **Array Creation:**
    *   `np.array()`: Used to create 1D (vectors), 2D (matrices), and higher-dimensional arrays (tensors) from Python lists.
    *   `dtype`: Allows explicit specification of data types for array elements (e.g., `int`, `float`, `str`, `bool`, `complex`).
    *   `np.arange()`: Creates arrays with evenly spaced values within a given range, with an optional step size.
    *   `np.linspace()`: Creates arrays with a specified number of evenly spaced values between a start and end point.

*   **Special Array Initialization:**
    *   `np.ones()`: Creates arrays filled with ones.
    *   `np.zeros()`: Creates arrays filled with zeros.
    *   `np.random.rand()`: Creates arrays with random values between 0 and 1.
    *   `np.identity()`: Creates an identity matrix (square matrix with ones on the main diagonal and zeros elsewhere).

*   **Array Manipulation:**
    *   `.reshape()`: Changes the shape (dimensions) of an array without altering its data, provided the total number of elements remains the same.
