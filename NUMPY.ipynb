```python
# IMPORT NUMPY AS NP
import numpy as np
```

```python
# import dataset 1D Array |  1D Array is also know as Vector
a = np.array([1,5,2,1])
print((a))
```

Q. Using numpy array in the front of list

A. Because perform all NumPy operations efficiently, use a NumPy array, not a regular Python list.

```python
# import dataset 2D Array | 2D Array is also know as Matrix
b = np.array([[1,5,6,2],[1,5,6,2]])
print(b)
```

```python
# import dataset 3D Array or Higher Array \ 3D Array or Higher Array is konw as tensor
c = np.array([[[1,1,1,1],[1,1,1,2]],[[1,1,1,3],[1,1,1,4]]])
print(c)
```

```python
# Data Type \ ( dtype )
A = np.array([5,9,6,3,5],dtype=int)
B = np.array([5,9,6,3,5],dtype=float)
C = np.array([5,9,6,3,5],dtype=str)
D = np.array([5,9,6,3,5],dtype=bool)
E = np.array([5,9,6,3,5],dtype=complex)
print(A)
print(B)
print(C)
print(D)
print(E)
```

see i can change the datatype of array to any datatype

```python
# Data Arange
A = np.arange(1,11)
print(A)
```

```python
B = np.arange(1,100,5)
print(B)
```

Cell 1 -> arange: Creates a NumPy array with evenly spaced values within a specified range.

Cell 2 -> it is written (1,100,5) The 5 after 100 means the gap between the numbers


```python
# Data Resape
A = np.arange(1,11).reshape(2,5)
print(A)
B = np.arange(1,13).reshape(3,4)
print(B)
```

Reshape : Changes the shape (dimensions) of a NumPy array without changing its data. it's vaild only reshape like (2,5),(3,5) like multiple vaild only
